### Creating the Tables

In this section, we will be creating the three tables of interest and populate them with the data from the CSV files already available to you. 
To get started, first upload the three CSV files to the DBFS as depicted in the following figure

![](https://i.imgur.com/QcCruBr.png)


Once you have done this, please remember to execute the following code to build the dataframes which will be saved as tables in our database

```
# File location and type
file_location_bookings = "/FileStore/tables/Bookings.csv"
file_location_facilities = "/FileStore/tables/Facilities.csv"
file_location_members = "/FileStore/tables/Members.csv"

file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
bookings_df = (spark.read.format(file_type) 
                    .option("inferSchema", infer_schema) 
                    .option("header", first_row_is_header) 
                    .option("sep", delimiter) 
                    .load(file_location_bookings))

facilities_df = (spark.read.format(file_type) 
                      .option("inferSchema", infer_schema) 
                      .option("header", first_row_is_header) 
                      .option("sep", delimiter) 
                      .load(file_location_facilities))

members_df = (spark.read.format(file_type) 
                      .option("inferSchema", infer_schema) 
                      .option("header", first_row_is_header) 
                      .option("sep", delimiter) 
                      .load(file_location_members))
```

####  Q7: How can you produce a list of all members who have used a tennis court?
- Include in your output the name of the court, and the name of the member formatted as a single column. 
- Ensure no duplicate data
- Also order by the member name.

```
%sql
select
    facilities.name as Facility_Name,
    concat(newbookings.firstname, ' ', newbookings.surname) as Member_Name
from (
    select distinct
        bookings.facid,
        members.memid,
        members.firstname,
        members.surname
    from members
    join bookings
    on members.memid = bookings.memid
    where members.memid <> 0
) as newbookings 
join facilities
on facilities.facid = newbookings.facid
where facilities.name like 'Tennis Court%'
order by
    newbookings.surname,
    facilities.name
;
```

```
spark.sql(
	'select '
	    'facilities.name as Facility_Name,'
	    'concat(newbookings.firstname, " ", newbookings.surname) as Member_Name '
	'from ('
	    'select distinct '
	        'bookings.facid,'
	        'members.memid,'
	        'members.firstname,'
	        'members.surname '
	    'from members '
	    'join bookings '
	    'on members.memid = bookings.memid '
	    'where members.memid <> 0'
	') as newbookings '
	'join facilities '
	'on facilities.facid = newbookings.facid '
	'where facilities.name like "Tennis Court%"'
	'order by '
	    'newbookings.surname,'
	    'facilities.name'
	';'
).show()
```

#### Q8: How can you produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30? 

- Remember that guests have different costs to members (the listed costs are per half-hour 'slot')
- The guest user's ID is always 0. 

#### Include in your output the name of the facility, the name of the member formatted as a single column, and the cost.

- Order by descending cost, and do not use any subqueries.

```
%sql
CONVERT TO DELTA bookings;
CONVERT TO DELTA members;
OPTIMIZE bookings ZORDER BY memid;
OPTIMIZE members ZORDER BY memid;
```

```
%sql
    select
        facilities.name,
        'Guest' as fullname,
        facilities.guestcost * bookings.slots as usage_cost
    from bookings
    join members
        on bookings.memid = members.memid
    join facilities
        on facilities.facid = bookings.facid
    where
        bookings.starttime like '2012-09-14%'
        and members.memid = 0
        and facilities.guestcost * bookings.slots > 30
union all
    select
        facilities.name,
        concat(members.firstname, ' ', members.surname) as fullname,
        facilities.membercost * bookings.slots as usage_cost
    from bookings
    join members
        on bookings.memid = members.memid
    join facilities
        on facilities.facid = bookings.facid
    where
        bookings.starttime like '2012-09-14%'
        and members.memid <> 0
        and facilities.membercost * bookings.slots > 30
order by
    usage_cost desc,
    facilities.name
;
```

```
spark.sql(
        "select "
            "facilities.name,"
            "'Guest' as fullname,"
            "facilities.guestcost * bookings.slots as usage_cost "
        "from bookings "
        "join members "
            "on bookings.memid = members.memid "
        "join facilities "
            "on facilities.facid = bookings.facid "
        "where "
            "bookings.starttime like '2012-09-14%'"
            "and members.memid = 0 "
            "and facilities.guestcost * bookings.slots > 30 "
    "union all "
        "select "
            "facilities.name,"
            "concat(members.firstname, ' ', members.surname) as fullname,"
            "facilities.membercost * bookings.slots as usage_cost "
        "from bookings "
        "join members "
            "on bookings.memid = members.memid "
        "join facilities "
            "on facilities.facid = bookings.facid "
        "where "
            "bookings.starttime like '2012-09-14%'"
            "and members.memid <> 0 "
            "and facilities.membercost * bookings.slots > 30 "
    "order by "
        "usage_cost desc,"
        "facilities.name"
    ";"
).show()
```

#### Q9: This time, produce the same result as in Q8, but using a subquery.

```
%sql
select
    newbookings.name as Facility_name,
    case
        when newbookings.memid = 0 then 'Guest'
        else concat(newbookings.firstname, ' ', newbookings.surname)
    end as Full_name,
    newbookings.usage_cost as Usage_cost
from (
    select
        bookings.bookid,
        facilities.name,
        members.firstname,
        members.surname,
        members.memid,
        case
            when members.memid = 0 then facilities.guestcost * bookings.slots
            else facilities.membercost * bookings.slots
        end as usage_cost
    from members
    join bookings
        on members.memid = bookings.memid
    join facilities
        on bookings.facid = facilities.facid
) as newbookings
join bookings
    on newbookings.bookid = bookings.bookid
where
    bookings.starttime like '2012-09-14%'
    and newbookings.usage_cost > 30
order by
    newbookings.usage_cost desc,
    newbookings.name
;
```

```
spark.sql(
    "select "
        "newbookings.name as Facility_name, "
        "case "
            "when newbookings.memid = 0 then 'Guest' "
            "else concat(newbookings.firstname, ' ', newbookings.surname) "
        "end as Full_name, "
        "newbookings.usage_cost as Usage_cost "
    "from ( "
        "select "
            "bookings.bookid, "
            "facilities.name, "
            "members.firstname, "
            "members.surname, "
            "members.memid, "
            "case "
                "when members.memid = 0 then facilities.guestcost * bookings.slots "
                "else facilities.membercost * bookings.slots "
            "end as usage_cost "
        "from members "
        "join bookings "
            "on members.memid = bookings.memid "
        "join facilities "
            "on bookings.facid = facilities.facid "
    ") as newbookings "
    "join bookings "
        "on newbookings.bookid = bookings.bookid "
    "where "
        "bookings.starttime like '2012-09-14%' "
        "and newbookings.usage_cost > 30 "
    "order by "
        "newbookings.usage_cost desc, "
        "newbookings.name "
    "; "
).show()
```

#### Q10: Produce a list of facilities with a total revenue less than 1000.
- The output should have facility name and total revenue, sorted by revenue. 
- Remember that there's a different cost for guests and members!

```
%sql
select
    facilities.name as Facility,
    sum(
        case
            when members.memid = 0 then facilities.guestcost * bookings.slots
            else facilities.membercost * bookings.slots
        end
    ) as TotalRevenue
from members
join bookings
    on members.memid = bookings.memid
join facilities
    on bookings.facid = facilities.facid
group by Facility
having TotalRevenue < 1000
order by TotalRevenue, Facility
;
```

```
spark.sql(
    "select "
        "facilities.name as Facility, "
        "sum( "
            "case "
                "when members.memid = 0 then facilities.guestcost * bookings.slots "
                "else facilities.membercost * bookings.slots "
            "end "
        ") as TotalRevenue "
    "from members "
    "join bookings "
        "on members.memid = bookings.memid "
    "join facilities "
        "on bookings.facid = facilities.facid "
    "group by Facility "
    "having TotalRevenue < 1000 "
    "order by TotalRevenue, Facility "
    "; "
).show()
```

```
%sql
select -- sums profits by month
    profitbymonth.facname as Facility,
    sum(profitbymonth.profit) as TotalProfit
from (
    select -- calculates profit as the sum of income in a month minus the corresponding monthly maintenance fee
        income.facname,
        sum(income.usage_cost) - income.monthlymaintenance as profit
    from (
        select -- gathers income per booking and corresponding monthly maintenance fee
            facilities.name as facname,
            facilities.monthlymaintenance,
            year(bookings.starttime) as year,
            month(bookings.starttime) as month,
            case
                when members.memid = 0 then facilities.guestcost * bookings.slots
                else facilities.membercost * bookings.slots
            end as usage_cost
        from members
        join bookings
            on members.memid = bookings.memid
        join facilities
            on bookings.facid = facilities.facid
    ) as income
    group by
        income.facname,
        income.monthlymaintenance,
        income.year,
        income.month
) as profitbymonth
group by profitbymonth.facname
having TotalProfit < 1000
order by
    TotalProfit,
    Facility
;
```
